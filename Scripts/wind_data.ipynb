```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import metpy.calc as mpcalc
from metpy.units import units

import warnings

from herbie import Herbie
```

```python


df = pd.read_csv("tmp_data/all_burned_wind_sites.csv")
df
```

```python
# Time in UTC, PDT 7 hrs ahead of UTC

xr.set_options(use_new_combine_kwarg_defaults=True)

i = []
for date in pd.date_range("2025-01-07 11:00:00", periods=9, freq="2h"):
    ds = Herbie(date).xarray("(?:U|V)GRD:10 m")
    i.append(
        ds.herbie.pick_points(df
            )
        )
slc_ts = xr.concat(i, dim="valid_time", compat='override')
```

```python
df
```

```python
df = slc_ts.to_dataframe()
```

```python
df['wdsp'] = mpcalc.wind_speed(df['u10'].values * units("m/s"), df['v10'].values * units("m/s")).magnitude

```

```python
df['wdir'] = mpcalc.wind_direction(df['u10'].values * units("m/s"), df['v10'].values * units("m/s")).magnitude
```

```python
df
```

```python

# Specify the file path
file_path = 'wind_data/all_hrrr.csv'

# Save the DataFrame to a CSV file
# index=False prevents writing the DataFrame index as a column in the CSV
df.to_csv(file_path, index=False)
```
